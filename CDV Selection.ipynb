```python
import pandas as pd
import random
```

```python
df_CDVExcel0 = pd.read_excel('Brandon Crouse CDV Request Q2 2021 01Jul2021.xlsx',sheet_name='Release')
df_CDVExcel1 = df_CDVExcel0[['Lot Type', 'MANUFACTURER']]

#creating unique drug column
dftemp=df_CDVExcel1['Lot Type'] 
Drug = [dftemp.str.split(' ')[index][0]
             for index in range(0, len(dftemp))]
df = pd.DataFrame(Drug, columns =['Drug'])
df['Drug']=df['Drug'].str.replace(r'\W',"")
df['Drug'] = df['Drug'].str.upper()
df_CDVExcel1['Drug']=df['Drug']

#Removing duplicate rows
df_CDVExcel2 = df_CDVExcel1.drop_duplicates()
df_CDVExcel2 = df_CDVExcel2.reset_index(drop=False)
df_CDVExcel2.to_excel('output4.xlsx', engine='xlsxwriter')
df_CDVExcel2.head()
```

```python
#Random selection function: N is the number of entries desired, output: the full randomly selecting N rows as output5.xlsx sheet

def random_selection(N):
    index = df_CDVExcel2.index
    if N > len(index):
        N = len(index)
    List0 = [random.randint(0,len(index)) for i in range(N)]
    temp = df_CDVExcel2.loc[List0,:]
    List1 = temp['index']
    Result = df_CDVExcel0.loc[List1,:]
    Result.to_excel('output5.xlsx', engine='xlsxwriter')

random_selection(10)
```

```python
#Random selection function2: 
# N is the number of entries desired
# D is the Drug variability desired in the random selection, 
    # 0: disregard variability, 
    # 1: Output all same drug (randomly selects a drug and if the drug entries are less than N than it will only output the available drug entries)
    # 2: Output all different drugs
# M is the Manufacturer variability desired in the random selection, 
    # 0: disregard variability, 
    # 1: Output all same manufacturer (randomly selects a manufacturer and if the manufacturer entries are less than N than it will only output the available drug entries)
    # 2: Output all different manufacturer
# Note: Either D or M has to be 0
# output: The full selecting rows as output5.xlsx sheet

def random_selection2(N, D, M):
    index = df_CDVExcel2.index
    if N > len(index):
        N = len(index)
    if D==0 and M==0:
        List0 = [random.randint(0,len(index)) for i in range(N)]
        temp = df_CDVExcel2.loc[List0,:]
        List1 = temp['index']
        Result = df_CDVExcel0.loc[List1,:]
        Result.to_excel('output5.xlsx', engine='xlsxwriter')
    if D==1 and M==0:
        dft1 = pd.DataFrame(df_CDVExcel1['Drug'].unique(), columns =['Unique Drug'])
        dfA=df_CDVExcel2.loc[df_CDVExcel2['Drug']==dft1.iloc[random.randint(0, len(dft1)),0]]
        if N > len(dfA):
            N = len(dfA)
            List1 = dfA['index']
            Result = df_CDVExcel0.loc[List1,:]
            Result.to_excel('output5.xlsx', engine='xlsxwriter')
        if N < len(dfA):
            dft2=dfA.sample(N)
            List1 = dft2['index']
            Result = df_CDVExcel0.loc[List1,:]
            Result.to_excel('output5.xlsx', engine='xlsxwriter')
    if D==2 and M==0:
        dft3 = df_CDVExcel2.drop_duplicates(subset='Drug', keep="first")
        dft4=dft3.sample(N)
        List1 = dft4['index']
        Result = df_CDVExcel0.loc[List1,:]
        Result.to_excel('output5.xlsx', engine='xlsxwriter')
    if D==0 and M==1:
        dft1 = pd.DataFrame(df_CDVExcel1['MANUFACTURER'].unique(), columns =['Unique MANUFACTURER'])
        dfA=df_CDVExcel2.loc[df_CDVExcel2['MANUFACTURER']==dft1.iloc[random.randint(0, len(dft1)),0]]
        if N > len(dfA):
            N = len(dfA)
            List1 = dfA['index']
            Result = df_CDVExcel0.loc[List1,:]
            Result.to_excel('output5.xlsx', engine='xlsxwriter')
        if N < len(dfA):
            dft2=dfA.sample(N)
            List1 = dft2['index']
            Result = df_CDVExcel0.loc[List1,:]
            Result.to_excel('output5.xlsx', engine='xlsxwriter')
    if D==0 and M==2:
        dft3 = df_CDVExcel2.drop_duplicates(subset='MANUFACTURER', keep="first")
        dft4=dft3.sample(N)
        List1 = dft4['index']
        Result = df_CDVExcel0.loc[List1,:]
        Result.to_excel('output5.xlsx', engine='xlsxwriter')
        
        
random_selection2(3, 2, 0)
```
